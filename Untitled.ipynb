```python
import pandas as pd
from sqlalchemy import create_engine 
```

```python
# Extract CSVs into DataFrames
draft_file = "Resources/basketball_draft (1).csv"
draft_df = pd.read_csv(draft_file)
draft_df.head() 
```

```python
# Create a filtered dataframe from specific columns
draft_filtered = ["draftYear", "draftRound", "draftSelection", "draftOverall", "tmID", "firstName", "lastName", "suffixName", "draftFrom"]
filtered_transformed = draft_df[draft_filtered].copy()

# Rename the column headers
filtered_transformed = filtered_transformed.rename(columns={"tmID": "TEAM_ABBREVIATION"})


# Clean the data by dropping duplicates and setting the index
# premise_transformed.drop_duplicates("id", inplace=True)
filtered_transformed.set_index("TEAM_ABBREVIATION", inplace=True)

filtered_transformed.head() 
```
