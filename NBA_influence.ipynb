```python
import pandas as pd
from sqlalchemy import create_engine 
```

```python
# Extract CSVs into DataFrames
draft_file = "Resources/basketball_draft (1).csv"
draft_df = pd.read_csv(draft_file)
draft_df.head() 
```

```python
# Filter years 2007 - 2017
draft_year_filtered = draft_df.loc[(draft_df['draftYear'] >=2000) & (draft_df['draftYear'] <= 2017)]
draft_year_filtered.head()
```

```python
# Create a filtered dataframe from specific columns
draft_filtered = ["draftYear", "draftRound", "draftSelection", "draftOverall", "tmID", "firstName", "lastName", "draftFrom"]
filtered_transformed = draft_year_filtered[draft_filtered].copy()

# Rename the column headers
filtered_transformed = filtered_transformed.rename(columns={"tmID": "TEAM_ABBREVIATION"})


# Clean the data by dropping duplicates and setting the index
# premise_transformed.drop_duplicates("id", inplace=True)
filtered_transformed.set_index("TEAM_ABBREVIATION", inplace=True)

filtered_transformed.head()
```

```python
# Merge firstName and lastName into a single column called PLAYER_NAME

filtered_transformed["player_name"] = filtered_transformed["firstName"] +" "+ filtered_transformed["lastName"]
filtered_transformed.set_index("player_name", inplace=True)
filtered_transformed.head()
```

```python
# Extract CSVs into DataFrames
nba_file = "Resources/nba_2016_2017_100.csv"
nba_df = pd.read_csv(nba_file)
nba_df.head() 
```

```python
# Create a filtered dataframe from specific columns
influence_cols = ["PLAYER_NAME", "TEAM_ABBREVIATION", "AGE", "W_PCT", "SALARY_MILLIONS", "TWITTER_FOLLOWER_COUNT_MILLIONS"]
influence_transformed = nba_df[influence_cols].copy()

# Clean the data by dropping duplicates and setting the index
# premise_transformed.drop_duplicates("id", inplace=True)
influence_transformed.set_index("PLAYER_NAME", inplace=True)

influence_transformed.head()
```

```python
# Create database connection
connection_string = "postgres:password@localhost:5432/nba_db"
engine = create_engine(f'postgresql://{connection_string}') 
```

```python
# Confirm tables
engine.table_names()
```

```python
# Load DataFrames into database
filtered_transformed.to_sql(name='draft', con=engine, if_exists='append', index=True) 
influence_transformed.to_sql(name='influence', con=engine, if_exists='append', index=True) 
```
